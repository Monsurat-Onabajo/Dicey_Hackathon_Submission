```python
# Check for columns that have missing values

def get_missing_columns(dataframe):
    """
    This function takes in a dataframe and returns columns that have null values in them
    
    Parameters
    -----------
    dataframe: (pd.DataFrame) dataset of the data you are working with.
    
    Returns
    --------
    Columns that have at least 1 missing value. 
    
    """
    missing_data= dataframe.isna().sum()
    missing_columns = missing_data[missing_data > 0]
    return missing_columns

fb= get_missing_columns(df_fb)
li= get_missing_columns(df_li)
in_= get_missing_columns(df_in)
tw= get_missing_columns(df_tw)

len(df_fb), len(df_tw), len(df_li), len(df_in)
```

### Derived Columns and Their Rationale:

1. **Year**:
    - **Type**: Integer
    - **Description**: Represents the year of the social media activity.
    - **Rationale**: Derived from the Date column to analyze year-over-year trends and assess the performance of posts across different years.

2. **Month**:
    - **Type**: Category
    - **Description**: Specifies the month of the social media activity.
    - **Rationale**: Useful for observing monthly patterns or seasonality in the data, aiding in understanding which months have higher or lower engagements.

3. **num of tags available**:
    - **Type**: Integer
    - **Description**: Indicates the number of tags a post has.
    - **Rationale**: Tags can significantly influence discoverability and engagement. This column helps in evaluating the impact of tags on post performance.

4. **Linked Content(?)**:
    - **Type**: Object (String)
    - **Description**: Determines if the post contains linked content.
    - **Rationale**: Linked content can drive traffic to external sites, making it vital to assess its presence and potential correlation with engagement.

5. **day_of_week**:
    - **Type**: Object (String)
    - **Description**: Denotes the day of the week of the post.
    - **Rationale**: Certain days might see higher engagements due to user activity patterns. This column aids in pinpointing optimal days for posting.

6. **is_weekend**:
    - **Type**: Integer (0 or 1)
    - **Description**: Indicates if the post was made on a weekend.
    - **Rationale**: Weekends might have different user activity patterns. This binary indicator helps in contrasting weekday vs. weekend performance.

7. **post_length**:
    - **Type**: Integer
    - **Description**: Gives the length of the post content in characters.
    - **Rationale**: Post length can influence readability and engagement. This metric allows for analysis of optimal post lengths.

8. **contains_question**:
    - **Type**: Integer (0 or 1)
    - **Description**: Highlights if the post has a question.
    - **Rationale**: Questions can boost engagement by prompting user responses. This column assesses the impact of questions on post interactions.

9. **num_hashtags**:
    - **Type**: Integer
    - **Description**: Counts the hashtags in the post.
    - **Rationale**: Hashtags can improve discoverability and categorization. Analyzing the number of hashtags helps in understanding their effect on post reach and engagement.

10. **part_of_day_posted**:
    - **Type**: Object (String)
    - **Description**: Categorizes the post by the time of day (morning, afternoon, evening).
    - **Rationale**: Posting time can influence visibility and engagement. This categorization aids in determining the best times to post for maximum impact.
    
11. **Positive_Reaction**:
    - **Type**: Integers
    - **Description**: Sum of 'Love Reactions', 'Haha Reactions', and 'Wow Reactions' in a single column
    - **Rationale**: It will help in determining the facebook data is performing positively

11. **Negative_Reaction**:
    - **Type**: Integers

## [Drop rows with negative values in the specified columns from the dataframe.](#Table-of-Contents)

```python
def drop_negative_rows(df, columns):
    """
    Drop rows with negative values in the specified columns from the dataframe.
    
    Parameters:
    - df: DataFrame containing the data.
    - columns: List of columns to check for negative values.
    
    Returns:
    - DataFrame with rows having negative values in the specified columns removed.
    """
    for col in columns:
        df = df[df[col] >= 0]
    return df

# List of columns to check for negative values
columns_to_check = ['Non-fan Organic Impressions', 'Other Post Clicks']
fb_df = drop_negative_rows(fb_df, columns_to_check)

# Check the shape of the cleaned data to confirm if any additional rows were dropped
fb_df.shape
```

```python
# Identify outliers in all numeric columns using the IQR method

# Select only numeric columns
numeric_data = fb_df.select_dtypes(include=['float64', 'int32'])

# Compute outliers for each numeric column
outliers = {}
for col in numeric_data.columns:
    Q1 = numeric_data[col].quantile(0.25)
    Q3 = numeric_data[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers[col] = ((numeric_data[col] < (Q1 - 1.5 * IQR)) | 
                     (numeric_data[col] > (Q3 + 1.5 * IQR))).sum()

outliers_df = pd.Series(outliers).reset_index()
outliers_df.columns = ['Column', 'Number of Outliers']

outliers_df.sort_values(by="Number of Outliers", ascending=False)
```

```python
def drop_negative_rows(df, columns):
    """
    Drop rows with negative values in the specified columns from the dataframe.
    
    Parameters:
    - df: DataFrame containing the data.
    - columns: List of columns to check for negative values.
    
    Returns:
    - DataFrame with rows having negative values in the specified columns removed.
    """
    for col in columns:
        df = df[df[col] >= 0]
    return df

# List of columns to check for negative values
columns_to_check = ['Engagements', 'Reactions', 'Likes', 'Comments', 'Shares']
li_df = drop_negative_rows(li_df, columns_to_check)
```

* I noticed a peak in impression in different years across different datasets and this makes me want to examine the correlation between Organic Impressions and total impressions across all datasets to ascertain if the company has ever opted for a paid impressions package in any of the social media platform.

```python
def plot_impressions(data, platform_name):
    plt.scatter(data["Impressions"], data["Organic Impressions"], alpha=0.7, label=platform_name)
    plt.xlabel("Total Impressions")
    plt.ylabel("Organic Impressions")
    plt.title(f"Organic Impressions vs Total Impressions for {platform_name}")
    plt.legend()
    plt.grid(False)
    plt.show()

# Plot with updated labels for each platform
plot_impressions(tw_df, "Twitter")
plot_impressions(in_df, "Instagram")
plot_impressions(li_df, "LinkedIn")
plot_impressions(fb_df, "Facebook")
```

 ## Key Insights

- A pronounced linear correlation exists between Total Impressions and Organic Impressions. This suggests that from 2018 to 2023, the company relied solely on organic reach without investing in paid impressions on any of these platforms.

* Check for the relationship between Part of the day the post was made and the post Average Engagements Across all social media platforms

```python
# Modify the function to return the grouped data
def get_avg_engagement_by_part_of_day(data, platform_name):
    # Group data by 'part_of_day_posted' and calculate the mean of 'engagements'
    avg_engagement = data.groupby("part_of_day_posted")["Engagements"].mean().reset_index()
    avg_engagement['Platform'] = platform_name
    return avg_engagement

# Get average engagement by part of day for each platform
tw_avg = get_avg_engagement_by_part_of_day(tw_df, "Twitter")
in_avg = get_avg_engagement_by_part_of_day(in_df, "Instagram")
li_avg = get_avg_engagement_by_part_of_day(li_df, "LinkedIn")
fb_avg = get_avg_engagement_by_part_of_day(fb_df, "Facebook")

# Concatenate all dataframes
combined_df = pd.concat([tw_avg, in_avg, li_avg, fb_avg])

# Define the color palette
colors = {'Twitter': '#E24197',
          'Instagram': '#FFD53F',
          'LinkedIn': '#324197',
          'Facebook': '#3C90A4'}

# Plot the combined data
plt.figure(figsize=(15, 8))
sns.barplot(data=combined_df, x="Engagements", y="part_of_day_posted", hue="Platform", ci=None, palette=colors)

# Set title and labels
plt.xlabel('Average Engagements')
plt.ylabel('Part of Day')
plt.title('Average Engagement by Part of Day Across Platforms')
plt.legend(title="Platform", loc='upper right')
plt.grid(axis='x')

plt.tight_layout()
plt.show()
```

## Observations
* Post made at night between 9pm-12am and in the morning between 5am-12pm usually tends to have higher engagements. 

## Insights
- Time of posting plays a crucial role in engagement. 
- Certain times of the day, such as nights and morning drive more interactions on all platforms.
- Understanding these peak times can aid in scheduling content for maximum impact.
